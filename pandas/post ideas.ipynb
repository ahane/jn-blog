### Adding rows to a DataFrame: The `concat` function

If you want to add entries to a DataFrame or combine two dataset that already have the same column-scheme you should use `concat`. It basically combines two DataFrames row-wise.

A `panads.concat()` example:

```python
books_nonfic = pd.DataFrame.from_csv('books/books_non_fiction.csv', index_col=None)
books_nonfic
```

```python
books_fic = pd.DataFrame.from_csv('books/books_fiction.csv', index_col=None)
books_fic
```

```python
all_books = pd.concat([books_fic, books_nonfic])
all_books
```

You can see that `concat()` takes a `list` of DataFrames as its argument. This can be any number of DataFrames.

Our resulting DataFrame preserved the (meaningless) indexing of our original DataFrames. If we don't care about the indices and instead want our new DataFrame enumerate afresh we can tell `concat` so:

```python
all_books = pd.concat([books_fic, books_nonfic], ignore_index=True)
all_books
```

### Combining rows from DataFrames with different columns

Lets take a look what happens if we try to concatenate DataFrames that don't have the same column-schema:

```python
toys = pd.DataFrame.from_csv('books/toys.csv', index_col=None)
toys
```

```python
products = pd.concat([books_fic, books_nonfic, toys], ignore_index=True)
products
```

Pandas tried its best and matched up the columns where it could. The new DataFrame contains all columns of original DataFrames had combined (the union of the columns). If a row came from a DataFrame where a column didn't exist, it gets the value `NaN` as you can see above. `NaN` stands for "Not a Number" and is pandas way of saying "missing value". How to work with missing values is a topic for another post.





### Outer and Inner Join of Columns

The union of the columns is sometimes called an 'outer join'. If we want our new DataFrame to instead only contain the columns that our DataFrames had in common (the intersection), we call this the 'inner join'

`concat` has a `join` argument to influence this behavior. It is set to `join='outer'` by default. The only other value it can take is `join='inner'`


```python
products = pd.concat([books_fic, books_nonfic, toys], ignore_index=True, join='inner')
products
```

Now we don't have any `NaN` values, as we only included the columns for which each row had a value.

```python
pd.concat
```

### The `axis` argument

By default `concat` combines DataFrames along the row-axis. Internally this row-axis (vertical, y-axis) is called the `axis 0`. 

The column-axis (horizontal, x-axis) is called `axis 1`. We can tell `concat` to combine along the column-axis: 

```python
products = pd.concat([books_fic, books_nonfic, toys], axis=1)
products
```

You can see that this does not really give useful results. The three DataFrames were just glued to each other side by side, without any regard for wether the rows are related. 

In some rare cases you might find yourself to want this naive glueing of Columns.

But Usually `concat` should not be used to combine DataFrames column-wise. For column-wise combinations you should use `join` and `merge`.
