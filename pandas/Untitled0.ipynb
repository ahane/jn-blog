The Pandas DataFrame
====================
In Excel or any other spreasheet application like GNumeric, etc. you are used to thinking in worksheets and workbooks. A worksheet is the big white empty table you see when you open the program. Three of these worksheets usually make up the workbook which you might therefore in its totality call 'the excel file'. When analyzing data in python, your basic concept for holding data is the DataFrame. DataFrame is pandas' equivalent of a single worksheet, and therefore basically a sophisticated table.

The first obvious difference to a worksheet is that the DataFrame is not immedeatly visible when created or opened.

```python
df = pd.DataFrame()
```

We just created our first DataFrame and gave it the name `df`. We can now take a look at `df` by typing just its name:

```python
df
```

You can see that it is quite different to an empty Excel file. It has 0 rows and 0 columns in contrast to the spreadsheet typical grid of alphabetical columns und numerated rows. You might have copy and pasted some data into your spreadsheet now. As this is not possible with pandas, lets look at a table that we load from a file. Almost any table like files will work with pandas. It is possible to load Excel xls and xlsx files, but for now it is easier to  look at csv files. 

CSV stands for 'comma seperated values' and is a textfile (\*.txt) based representations of simple tables. The first row of a basic text file lists the names of the columns which are seperated by commas. The following rows represent one datapoint and list their values again separated by commas. Many Excel worksheets can be directly saved as csv files. Often though the worksheet contains additional cells that show additional text or calculated values. Then the homogenous part of the workesheet, the data itself can be easily selected and copy-pasted into another clean worksheet. Lets load such an csv file:

```python
finance_industry = pd.DataFrame.from_csv('datasets/us_finance_industry.csv')
```

This time we chose a more significant name for our DataFrame, although in many cases you might be happy with a conscise `df`.
Lets look at what we loaded:

```python
finance_industry
```

Pandas shows us the first 60 rows of our table, but at the bottom you can read that our data actually contains 84 rows. This is fine, most of the time there is no need for looking at all entries/rows of a table individually. I is enough to know what columns exist and what a typical value for such a column might look like. We can look at slices of rows by adding `.ix[]` (short for index) to the name of our DataFrame and putting row-ranges in the brackets:

```python
finance_industry.ix[0:5]
```

We can also look at just some columns by adding a list of column names in brackets (One set of brackets to create a list, one of brackets as the syntax for selecting columns):

```python
finance_industry[['year', 'Domestic']]
```

And of course we can combine the two:

```python
finance_industry.ix[0:5][['year', 'Domestic']]
```

This interactive way of looking at just parts of the DataFrame is what makes pandas data analysis so powerful. It has multiple advantages:
- Our tables can be huge, up to multiple millions of rows. Try to image Excel loading a 300mb file!
- We don't select cell ranges with the mouse. Instead we usually handle indiviual or multiple columns, and sometimes rows. This makes it almost impossible to accidentally exlude data.
- Multiple of these digestable, sliced and diced representations of our original dataset together can make up the data story of our notebook at hand. They can be annotated with markdown text and thus explain how certain conclusions about the data have been reached.

For the rest of this article we shall look only at three columns:


```python
profits = finance_industry[['year', 'CorporateProfitsAdj', 'Financial']]
```

```python
profits.ix[:5]
```

By now you might have noticed that the names of our columns are little text bits (strings) instead of the letters used by Excel. The actualy meaning of the column is usually put in one of the top rows of our spreasheet. In pandas the meaning of the column and the name of the column are the same. They can be easily changed by assinging a new list of strings to the `.columns` value of our DataFrame:

```python
profits.columns = ['year', 'total', 'finance']
profits.ix[:5]
```

By default the names of the rows are numbers like in Excel. The names of the rows are refered to as the `index` of the DataFrame. We can designate one of our columns to be the index instead of the default numbers:

```python
profits.index = profits['year']
profits.ix[1990:2000]
```

We can now slice our DataFrame by years. This is indeed most useful when working with timeseries data. It can also power sophisticated data transformations when used with hierarchical indexes, but we are getting ahead of ourselves.

## Colums and Rows vs Cells and Cell Ranges

You might have noticed that so far we have only looked at multiple columns and multiple multiple rows at once. We can also look at a single column typing just one name into the columns and omitting the second set of braces: 

```python
profits['total']
```

You can see that the representation of our column now looks quite different. This is because a single column makes up what pandas calls a `Series`. `Series` behave somewhat differently then DataFrames, but for now it doesn't really matter in what ways. We will keep thinking in terms of columns as it is much more intuitive. 

It is important to understand how fundamental `Series` are to `pandas`. They are the basic concept for working with our data. Everything is done in aggregate, over all data points.

In Excel you might work with a single cell in one moment and with a cell range (that might span a row, a column or an area) in the next. In `pandas` it is always about whole `Series`; a whole column (a cell range over a column `A1:A100`). If you find yourself wanting to handle individual cells or even rows, you should stop and try to find a way to express your desired result though the use of columns (and row-ranges of columns).


## Formular Bar and Python Variables

Probabily the most common operation in Excel is clicking on the Formular Bar, typing `=` followd by some formula and looking at the computed value in the cell that we had selected before. Additionally we might drag the Fill Handle [IMG] down a column or across some rows to apply the formular to the respective other cells.

In pandas the whole process is quite different. We always apply "formulas", now called functions, to whole columns. This might seem like a restriction at first but has incredible speed benefits because of the way `pandas` was built. 


### Computing a Formula over a cell range

Let assume we want to compute the sum of the total profits throughout all years:
In Excel this might look like this: `D20=SUM(A1:A89)`

In pandas it looks like this:

```python
total_all_years = profits['total'].apply(sum)
```

Like Excel, `pandas` has many inbuilt functions, of which `sum()` is one. We can easily define our own functions, or import them from other python programs (libraries). Because python is a fully-featured programming language, there are literally no limits to what these functions can do. We will look into how to replicate some common Excel formulas in a later post.

Note that instead of saving our computed single (scalar) value into a cell, we saved it into a Python variable. We can retriefe the value later by just typing the variables name:

```python
total_all_years
```

What we just did was equivalent to selecting a cell and computing a formula over a (column-long) cell-range. 



### Dragging a Formula over multiple cells


Another common thing to do is select a cell compute a formula from a cell in the same row and then drag the Fill Handle to apply the formula to a whole column. 

Assume we want to convert the profits from billions into millions. 

To do so we have to multiply each value with `1000` In Excel this would look like this: `B1=A1*1000` dragged over as long as our column is. In pandas we simply type this:

```python
profits['total_mio'] = profits['total']*1000
```

Its okay if this reads a bit confusing at first. Lets take a look at our DataFrame and see what has happened:

```python
profits.ix[1990:1995]
```

By typing `profits['total_mio']` the left side of the equal sign, we managed to directly add the result of our computation to our DataFrame. This on the fly creation of new columns is very common in pandas. Often coming to a desired result will involve creating multiple intermediate columns first. If we create a column on accident or don't need it any more we can delete it with the python command `del`:


```python
del profits['total_mio']
```

The next confusing bit might be `profits['total']*1000`. We multiplied the whole column with 1000. What does this mean? 
You can think of it like applying a function to a column again, think of it as a function like `profits['total'].apply(multiply_by_100)`. The difference to above then is not in the notation. It is in the fact that the function was applied to each row in the column and then all new values where returned together as a new column. Instead of getting a single (scalar) value from our function we received a column again.

This is what I ment by "you should think in columns". Columns become the basic thinking block in your math. You are calculating with columns, you are doing vecor math! Yes that's right you are doing advanced algebra, without having to worry about the details.

Lets try to compute the percentual share of finance in in the total profits:

```python
profits['finance_ratio'] = profits['finance']/profits['total']
profits['finance_percent'] = profits['finance_ratio']*100
profits[['finance_percent']].ix[1990:1999]
```
